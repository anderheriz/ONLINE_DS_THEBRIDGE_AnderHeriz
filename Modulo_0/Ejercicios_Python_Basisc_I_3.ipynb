## Ejercicio 3
Mismo programa que en 2, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
valor1=input("Introduce un valor: ")
valor2=input("Introduce segundo valor: ")
valor3=input("Introduce tercer valor:")

if valor1==valor2==valor3:
  print("son iguales")
else:
  print("no son iguales")

if valor1==valor2 or valor1==valor3 or valor2==valor3:
  print("al menos dos son iguales")
else:
  print("todos los valores son diferentes")
````
