Haz un programa que detecte todos los duplicados de un elemento en una lista:

* El programa debe devolver todos los índices de los elementos duplicados.
* Aplícalo para encontrar los duplicados de "un", "es" y "binario" de:  

>```ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]```


`````python
ejer_3 = ["Un", "árbol", "binario", "es", "una", "estructura", "de", "un", "tipo", "particular", "a", "veces", "no", "es", "ni", "binario"]
def encontrar_duplicados(lista,duplicado):
    indices=[]
    for i, valor in enumerate(lista):
        if valor.lower==duplicado.lower:
            indices.append(i)
    return indices
print("Duplicado 'un':",encontrar_duplicados(ejer_3,"un"))
print("duplicado 'es':",encontrar_duplicados(ejer_3,"es"))
print("duplicado 'binario':",encontrar_duplicados(ejer_3,"binario"))
`````

## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
def dni_valido(dni):
  palabra='TRWAGMYFPDXBNJZSQVHLCKE'
  if len(dni)!=9:
    return False
  numero=dni[:-1]
  letra=dni[-1]
  if not numero.isdigit():
    return False
  if len(numero)<7 or len(numero)>8:
    return False
  letra_correcta= palabra[int(numero)%23]
  return letra== letra_correcta
dni=input("Introduce DNI:")
if dni_valido(dni):
  print("Dni valido")
else:
  print("DNI no valido")
`````
